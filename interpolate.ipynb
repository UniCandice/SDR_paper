```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import PchipInterpolator

# Load the data
data = np.loadtxt('/home/campus.ncl.ac.uk/nll93/Documents/code_saturne_simulations/Vflame-temp_keLS/reac/k_epsilon/DATA/k.txt')

# Extract x and y values
x = data[:, 0]
y = data[:, 1]

# Ensure y values are strictly increasing and within the range [0, 2]
if not np.all(np.diff(y) > 0):
    raise ValueError("y values are not strictly increasing.")

if not (np.min(y) >= 0 and np.max(y) <= 2):
    raise ValueError("y values are not in the range [0, 2].")

# Define the new resolution for y
m = 400
yi = np.linspace(y.min(), y.max(), m)

# Perform PCHIP interpolation based on y values
pchip = PchipInterpolator(y, x)
xi = pchip(yi)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o', label='Original data (98 points)')
plt.plot(xi, yi, '-', label='Interpolated data (400 points)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('kenetic Interpolation based on y')
plt.show()

# Save the interpolated data to a file
np.savetxt('interpolated_k_based_on_y.txt', np.column_stack((xi, yi)), fmt='%.8f', header='x y')

```
